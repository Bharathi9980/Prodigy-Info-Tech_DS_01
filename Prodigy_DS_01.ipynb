```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np


gender_data = ['Male', 'Female', 'Male', 'Female', 'Female', 'Male', 'Female', 'Male', 'Male', 'Female',
               'Male', 'Female', 'Male', 'Male', 'Female', 'Female', 'Male', 'Female', 'Male', 'Female']


age_data = np.random.randint(18, 60, size=100)


plt.figure(figsize=(10, 5))


plt.subplot(1, 2, 1)
sns.countplot(x=gender_data, palette='Set2')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')


plt.subplot(1, 2, 2)
plt.hist(age_data, bins=10, edgecolor='black', color='skyblue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')


plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd


data = {'Gender': ['Male', 'Female', 'Male', 'Female', 'Female', 'Male', 'Female', 'Male', 'Male', 'Female',
                   'Male', 'Female', 'Male', 'Male', 'Female', 'Female', 'Male', 'Female', 'Male', 'Female'],
        'Age Group': ['18-25', '18-25', '26-35', '26-35', '36-45', '46-55', '46-55', '56+', '56+', '18-25',
                      '18-25', '26-35', '26-35', '36-45', '46-55', '46-55', '56+', '56+', '18-25', '26-35']}

df = pd.DataFrame(data)


pd.crosstab(df['Age Group'], df['Gender']).plot(kind='bar', stacked=True, color=['skyblue', 'orange'])
plt.title('Gender Distribution Across Age Groups')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np


age_data = np.random.randint(18, 60, size=100)


sns.kdeplot(age_data, shade=True, color='purple')
plt.title('Age Distribution (KDE)')
plt.xlabel('Age')
plt.ylabel('Density')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


age_data_group1 = np.random.randint(18, 25, size=100)
age_data_group2 = np.random.randint(26, 35, size=100)
age_data_group3 = np.random.randint(36, 45, size=100)


plt.figure(figsize=(12, 6))


plt.hist(age_data_group1, bins=7, alpha=0.5, label='18-25', color='blue', edgecolor='black')


plt.hist(age_data_group2, bins=7, alpha=0.5, label='26-35', color='green', edgecolor='black')


plt.hist(age_data_group3, bins=7, alpha=0.5, label='36-45', color='red', edgecolor='black')

plt.title('Age Distribution Across Different Groups')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
